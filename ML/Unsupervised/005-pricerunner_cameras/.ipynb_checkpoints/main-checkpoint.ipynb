```python
import pandas as pd
import numpy as np
data = pd.read_csv('pricerunner_cameras.csv')
data.head()
```

```python
data = data.drop(data.iloc[:, 0:1], 1)
```

```python
from sklearn.preprocessing import LabelEncoder
data = data.apply(LabelEncoder().fit_transform)
data.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for i in range(1, 16):
    kmeans = KMeans(i, n_init=15)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
    
plt.plot(wcss, 'bx-')
plt.show()
```

```python
kmeans = KMeans(2)
Y_kmeans = kmeans.fit_predict(data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(2)
x = pca.fit_transform(data)
```

```python
x = data.values
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x[Y_kmeans==0, 0], x[Y_kmeans==0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(x[Y_kmeans==1, 0], x[Y_kmeans==1, 1], s=50, c='green', label='Cluster 2')
plt.scatter(x[Y_kmeans==2, 0], x[Y_kmeans==2, 1], s=50, c='blue', label='Cluster 3')
plt.legend()
plt.show()
```
