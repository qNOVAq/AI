```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
plt.style.use('ggplot')
warnings.filterwarnings('ignore')

data = pd.read_csv('data/diabetes.csv')
data.head()
```

```python
data.describe().T
```

```python
#Checking for missing values
data.isnull().sum()
```

```python
#A function to make plotting easier
def plot(x, color, title):
    plt.figure(figsize=(12,10))
    sns.distplot(x, color=color).set_title(title)
```

```python
plot(data.Pregnancies, 'red', 'Distribution of Pregnancies')
```

```python
plot(data.Glucose, 'green', 'Distribution of Glucose')
```

```python
plot(data.BloodPressure, 'blue', 'Distribution of Blood Pressure')
```

```python
plt.figure(figsize=(16,14))
sns.barplot(data.Age, data.Outcome).set_title('Effect of Age on Outcome')
```

```python
#Countplot of the dependant variable
plt.figure(figsize=(12,10))
sns.countplot(data.Outcome).set_title('Count of the Outcome')
```
