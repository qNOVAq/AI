```python
# Droppping low correlated column
data = data.drop(['Longtitude', 'Lattitude', 'Address', 'Date', 'Postcode', ], 1)
```

```python
# Basic featrure engineering
data['Total_Rooms'] = data['Rooms'] + data['Bedroom2'] /2
data['Bathroom per bedroom'] = data['Bedroom2'] / data['Bathroom']
```

```python
# Filling missing values with their mean and median
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy='most_frequent')

data['Price'] = data['Price'].fillna(data['Price'].mean())
data['Distance'] = data['Distance'].fillna(data['Distance'].mean())
data['Bedroom2'] = data['Bedroom2'].fillna(data['Bedroom2'].mean())
data['Bathroom'] = data['Bathroom'].fillna(data['Bathroom'].mean())
data['Car'] = data['Car'].fillna(data['Car'].mean())
data['Landsize'] = data['Landsize'].fillna(data['Landsize'].mean())
data['BuildingArea'] = data['BuildingArea'].fillna(data['BuildingArea'].mean())
data['YearBuilt'] = data['YearBuilt'].fillna(data['YearBuilt'].mean())
data['Propertycount'] = data['Propertycount'].fillna(data['Propertycount'].mean())
data['Total_Rooms'] = data['Total_Rooms'].fillna(data['Total_Rooms'].mean())
data['Bathroom per bedroom'] = data['Price'].fillna(data['Price'].mean())

data.iloc[:, 13:14] = si.fit_transform(data.iloc[:, 13:14])
data.iloc[:, 14:15] = si.fit_transform(data.iloc[:, 14:15])

data.isnull().sum()
```

```python
# Encoding categorical variables
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

categorical = data.select_dtypes(include='object')
categorical_cols = categorical.columns

for column in categorical_cols:
    data[column] = encoder.fit_transform(data[column])
```

```python
# Converting non float to float64
non_float = data.select_dtypes(exclude='float64')
non_float_cols = non_float.columns

for column in non_float_cols:
    data[column] = data[column].astype('float64')
data.info()
```

```python
# Splitting to dependant and independant
X = data.drop('Price', 1).values
y = data.Price.values
```

```python
# Splitting to train and test/validation sets
from sklearn.model_selection import KFold
kf = KFold()

for train_index, test_index in kf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = y[train_index], y[test_index]
    
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'Y_train shape: {Y_train.shape}')
print(f'X_test shape: {Y_test.shape}')
```

```python
# Normalizing the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting to the model
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

estimators = [LinearRegression(), RandomForestRegressor(), DecisionTreeRegressor(),
             XGBRegressor()]

for model in estimators:
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)

    accuracy = model.score(X_test, Y_test)
    
    mse = mean_absolute_error(Y_test, Y_pred)
    print(f'{model} Accuracy: {accuracy}')
    print(f'{model} Error: {mse}')
    print('='*100)
```
