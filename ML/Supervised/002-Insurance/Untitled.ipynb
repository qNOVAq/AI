```python
import pandas as pd
import numpy as np

data = pd.read_csv('insurance/insurance.csv')
cols = data.columns
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder_1 = LabelEncoder()
data.sex = encoder_1.fit_transform(data.sex)

encoder_2 = LabelEncoder()
data.smoker = encoder_2.fit_transform(data.smoker)
```
