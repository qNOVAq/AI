```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')

data = pd.read_csv('data/vgsales.csv')
data.head()
```

```python
plt.figure(figsize=(10, 9))
sns.heatmap(data.corr(), annot=True)
```

```python
data.info()
```

```python
#Filling missing values with most frequent
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy='most_frequent')
data.iloc[:, 3:4] = si.fit_transform(data.iloc[:, 3:4])

data.iloc[:, 5:6] = si.fit_transform(data.iloc[:, 5:6])

data.isnull().sum()
```

```python
data.head()
```

```python
plt.figure(figsize=(20, 13))
sns.countplot(data['Platform'], saturation=1)
```

```python
#Top 3 platform's global sales
plt.figure(figsize=(15,10))
plt.subplot(331)
sns.barplot(x=data[data.iloc[:, 3]=='DS'], y='Global_Sales', data=data)

plt.subplot(332)
sns.barplot(x=data[data.iloc[:, 3]=='PS2'], y='Global_Sales', data=data)

plt.subplot(333)
sns.barplot(x=data[data.iloc[:, 3]=='PS3'], y='Global_Sales', data=data)
```
