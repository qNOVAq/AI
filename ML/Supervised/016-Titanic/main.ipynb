```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('fivethirtyeight')

data_train = pd.read_csv('data/train.csv')

print(f'Training shape: {data_train.shape}')
```

```python
print(f'Training set missing values: {data_train.isnull().sum()}')
```

```python
from sklearn.impute import SimpleImputer

si = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

train_cols = data_train.columns

data_train = si.fit_transform(data_train)

data_train = pd.DataFrame(data_train, columns=train_cols)
```

```python
print(f'Training set missing values: {data_train.isnull().sum()}')
```

```python
#dropping low correlated features
dropped_columns = ['Name', 'Ticket', 'Cabin', 'PassengerId']

data_train = data_train.drop(dropped_columns, 1)

data_cols = data_train.columns
```

```python
categorical_columns =  data_train.select_dtypes(include='object').columns

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for column in categorical_columns:
    #training set
    data_train[column] = encoder.fit_transform(data_train[column])
    
data_train = pd.DataFrame(data_train, columns=data_cols)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data_train.corr(), annot=True)
```

```python
from sklearn.preprocessing import StandardScaler

x = data_train.drop('Survived', 1)
y = data_train.Survived

sc = StandardScaler()

x = sc.fit_transform(x)
```

```python
x = pd.DataFrame(x, columns=data_cols[1:])

plt.figure(figsize=(5,6))
sns.countplot(y)
```

```python
plt.figure(figsize=(20,18))
plt.subplot(331)
sns.distplot(x.Age, color='red', bins=50)

plt.subplot(332)
sns.distplot(x.Fare, color='blue', bins=50)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
import warnings 
warnings.filterwarnings('ignore')
estimators = [LogisticRegression(), SVC(), LinearSVC(), RandomForestClassifier(), KNeighborsClassifier(), GaussianNB(),DecisionTreeClassifier()]

for model in estimators:
    scores = cross_val_score(model, x,y, cv=8)
    print(f'{model} Accuracy: {scores.mean()}')
    print('-'*40)
```
