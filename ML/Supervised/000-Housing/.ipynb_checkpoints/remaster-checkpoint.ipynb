```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
plt.style.use('fivethirtyeight')
warnings.filterwarnings('ignore')

# Loading the data
data = pd.read_csv('data/housing.csv')
data.head()
```

```python
data.describe().T
```

```python
# Checking for missing values
data.isnull().sum()
```

```python
# Filling missing values with the most frequent
from sklearn.impute import SimpleImputer
si = SimpleImputer(np.nan, strategy='most_frequent')

data.iloc[:, 4:5] = si.fit_transform(data.iloc[:, 4:5])

# Checking for missing values
data.isnull().sum()
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(data.corr(), annot=True).set_title('Correlation heat map of the data')
```

```python
# dropping longitude and latitude
data = data.drop(['longitude', 'latitude'], 1)
```

```python
# Basic feature engineering
data['population_per_household'] = data['population'] / data['households']
data['median_income_per_household'] = data['median_income'] / data['households']
```

```python
# Simple EDA
plt.figure(figsize=(15,12))
sns.distplot(data.population, color='red').set_title('Distribution of Population')
```

```python
plt.figure(figsize=(15,12))
sns.distplot(data.median_income, color='green').set_title('Distribution of Median Income')
```

```python
plt.figure(figsize=(15,12))
sns.distplot(data.median_house_value, color='blue').set_title('Distribution of the Median House Value')
```

```python
plt.figure(figsize=(15,20))
sns.barplot(data.ocean_proximity, data.median_income).set_title('Income in different areas')
```

```python
# Encoding vategorical columns 
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

data['ocean_proximity'] = encoder.fit_transform(data['ocean_proximity'])
data['ocean_proximity'] = data['ocean_proximity'].astype('float64')
```

```python
# Splitting to dependant and indepandant 
X = data.drop('median_house_value', 1).values
y = data.median_house_value.values
```

```python
# Splitting to train and test set
from sklearn.model_selection import KFold
kf = KFold()

for train_index, test_index in kf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = y[train_index], y[test_index]
    
print(f'X_train shape: {X_train.shape}')
print(f'Y_train shape: {Y_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'Y_test shape: {Y_test.shape}')
```

```python
# Normalize the data
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from tqdm import tqdm

estimators = [LinearRegression(), DecisionTreeRegressor(),SVR(),
              XGBRegressor(), RandomForestRegressor()]
for model in tqdm(estimators):
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    
    accuracy = model.score(X_test, Y_test)
    mse = mean_squared_error(Y_test, Y_pred)
    
    print(f'{model} Accuracy: {accuracy}')
    print(f'{model} Error: {mse}')
    print('='*100)
    
```
