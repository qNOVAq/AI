```python

import pandas as pd
import numpy as np
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
```

```python
data = pd.read_csv('heart.csv')
data.head()
```

```python
#missing values
data.isnull().sum()
```

```python
x = data.drop('target', 1)
y = data.target
```

```python
#preprocessing
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import cross_val_score
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
ann = Sequential()

def build_model():
    
    
    ann.add(Dense(20, input_dim=x.shape[1], activation='relu'))
    Dropout(.2)
    
    ann.add(Dense(13, activation='relu'))
    Dropout(0.2)
    
    
    ann.add(Dense(1, activation='sigmoid'))
    
    ann.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
    
    #Saving weights
    ann.save_weights('weights.h5')
    return ann

classifer = KerasClassifier(build_model, epochs=250, batch_size=32)
scores = cross_val_score(classifer, x, y, cv=3)
```
