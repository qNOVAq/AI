```python
import numpy as np
from tensorflow.keras.layers import Dense, Dropout, MaxPooling2D, Conv2D, BatchNormalization, Flatten
from tensorflow.keras.models import Sequential
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
import cv2
```

```python
train_dir = 'seg_train/seg_train'
test_dir = 'seg_test/seg_test'

train_gen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2,
                               horizontal_flip=True)
training_set = train_gen.flow_from_directory(train_dir, target_size=(64,64),
                                             class_mode='categorical', batch_size=32)

test_gen = ImageDataGenerator(rescale=1./255)
testing_set = test_gen.flow_from_directory(test_dir, target_size=(64,64)),
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau
rl = ReduceLROnPlateau(monitor='val_loss', cooldown=0, patience=4, verbose=1, factor=0.1)
```

```python
cnn = Sequential()

cnn.add(Conv2D(32, input_shape=[64,64,3], activation='relu', kernel_size=3))
cnn.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))

cnn.add(Conv2D(64, activation='relu', kernel_size=3))
cnn.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))
cnn.add(Dropout(0.5))

cnn.add(Flatten())

cnn.add(Dense(128, activation='relu'))
cnn.add(Dropout(0.2))

cnn.add(Dense(6, activation='softmax'))

cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = cnn.fit(x=training_set, validation_data=testing_set, epochs=30, callbacks=[rl])
```

```python
cnn.save_weights('weights.h5')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(25,15))

plt.subplot(331)
plt.plot(cnn.history.history['accuracy'], 'bo--', label='Accuracy',c='red')
plt.plot(cnn.history.history['val_accuracy'], 'bo--', label='Validation Accuracy',c='blue')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('train_acc vs val_acc')
plt.legend()

plt.subplot(332)
plt.plot(cnn.history.history['loss'], 'bo--', label='Loss',c='red')
plt.plot(cnn.history.history['val_loss'], 'bo--', label='Val_Loss',c='blue')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('train_loss vs val_loss')
plt.legend()
plt.show()
```
